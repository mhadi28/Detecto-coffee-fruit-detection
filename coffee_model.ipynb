```python
from os import listdir
```

```python
path = '/content/drive/MyDrive/Dataset detecto'
listdir(path)
```

```python
import matplotlib.pyplot as plt
from detecto.utils import read_image
```

```python
from detecto.core import Dataset, DataLoader, Model
from detecto.visualize import show_labeled_image

dataset = Dataset(path)
image, targets = dataset[0]
show_labeled_image(image, targets['boxes'], targets['labels'])
```

```python
labels = ['coffee_fruit']
model = Model(labels)
```

```python
import torch 
if torch.cuda.is_available():  
  dev = "cuda:0" 
else:  
  dev = "cpu"
```

```python
model.fit(dataset)
```

```python
model.save('coffee_fruit.pth')
```

```python
image = read_image('/content/drive/MyDrive/Dataset detecto/image (36).jpg')
predictions = model.predict(image)

# predictions format: (labels, boxes, scores)
labels, boxes, scores = predictions

print(labels) 
print(boxes)
print(scores)
```

```python
from detecto import visualize
visualize.show_labeled_image(image, boxes, labels)
```
